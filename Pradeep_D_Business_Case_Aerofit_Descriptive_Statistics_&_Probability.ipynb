### Contingency Table

```python
#Contingency Table b/w the Product and the Gender.

ct = pd.crosstab(df['Product'], df['Gender'])
print(ct)
```

**Insights**
* KP281 has been equally purchased by both males and females (40 each), indicating no gender preference for this product.
* KP481 shows a slight preference among males (31 males vs 29 females), though the difference is minimal.
* KP781 is significantly more popular among males (33 males vs 7 females), suggesting a strong gender-based preference.




```python
# Finding the marginal probablity of the products.

marginal_p_KP281 = ct.loc['KP281'].sum()/ct.values.sum()*100
print(f"Marginal Probablity of product KP281 is {round(marginal_p_KP281,2)}%")

marginal_p_KP481 = ct.loc['KP481'].sum()/ct.values.sum()*100
print(f"Marginal Probablity of product KP481 is {round(marginal_p_KP481,2)}%")

marginal_p_KP781 = ct.loc['KP781'].sum()/ct.values.sum()*100
print(f"Marginal Probablity of product KP781 is {round(marginal_p_KP781,2)}%")
```

```python
ct
```

**Insights**
* 44.4% of customers have purchased KP281, making it the most popular product.
* 33.3% of customers have purchased KP481, showing moderate popularity.
* 22.2% of customers have purchased KP781, the least purchased product.

```python
# Finding the conditional probablity of the products.

print("For KP281:")
#Conditional probablity P(KP281 | Male)

kp281_male = ct.loc['KP281', 'Male']/ct['Male'].sum()*100
print(f"Probablity of Male customer buying product KP281 is {round(kp281_male,2)}%")

#Conditional probablity P(KP281 | Female)

kp281_female = ct.loc['KP281', 'Female']/ct['Female'].sum()*100
print(f"Probablity of Female customer buying product KP281 is {round(kp281_female,2)}%")

#Conditional probablity P(Male | KP281)

kp_male_281 = ct.loc['KP281', 'Male']/ct.loc['KP281'].sum()*100
print(f"Probablity of Male customer given that product is KP281 is {round(kp_male_281,2)}%")

#Conditional probablity P(Female | KP281)

kp_female_281 = ct.loc['KP281', 'Female']/ct.loc['KP281'].sum()*100
print(f"Probablity of Female customer given that product is KP281 is {round(kp_female_281,2)}%\n")



print("For KP481:")
#Conditional probablity P(KP481 | Male)

kp481_male = ct.loc['KP481', 'Male']/ct['Male'].sum()*100
print(f"Probablity of Male customer buying product KP481 is {round(kp481_male,2)}%")

#Conditional probablity P(KP481 | Female)

kp481_female = ct.loc['KP481', 'Female']/ct['Female'].sum()*100
print(f"Probablity of Female customer buying product KP481 is {round(kp481_female,2)}%")

#Conditional probablity P(Male | KP481)

kp_male_481 = ct.loc['KP481', 'Male']/ct.loc['KP481'].sum()*100
print(f"Probablity of Male customer given that product is KP481 is {round(kp_male_481,2)}%")

#Conditional probablity P(Female | KP481)

kp_female_481 = ct.loc['KP481', 'Female']/ct.loc['KP481'].sum()*100
print(f"Probablity of Female customer given that product is KP481 is {round(kp_female_481,2)}%\n")



print("For KP781:")
#Conditional probablity P(KP781 | Male)

kp781_male = ct.loc['KP781', 'Male']/ct['Male'].sum()*100
print(f"Probablity of Male customer buying product KP781 is {round(kp781_male,2)}%")

#Conditional probablity P(KP781 | Female)

kp781_female = ct.loc['KP781', 'Female']/ct['Female'].sum()*100
print(f"Probablity of Female customer buying product KP781 is {round(kp781_female,2)}%")

#Conditional probablity P(Male | KP781)

kp_male_781 = ct.loc['KP781', 'Male']/ct.loc['KP781'].sum()*100
print(f"Probablity of Male customer given that product is KP781 is {round(kp_male_781,2)}%")

#Conditional probablity P(Female | KP781)

kp_female_781 = ct.loc['KP781', 'Female']/ct.loc['KP781'].sum()*100
print(f"Probablity of Female customer given that product is KP781 is {round(kp_female_781,2)}%\n")
```
